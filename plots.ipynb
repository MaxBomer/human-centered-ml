```python
# Function that plots mean accuracy-per-round-curve across (3) runs

def plot_accuracy_curves(dataset_name, results, figsize=(10,6)):

    dataset = results[dataset_name]

    plt.figure(figsize=figsize)


    for strategy, entries in dataset.items():
        accuracies = entries['accuracies']
        num_rounds = len(accuracies[0])
        rounds = np.arange(num_rounds)

        if len(accuracies) > 1:
            stacked = np.stack(accuracies)
            mean_curve = stacked.mean(axis=0)
            std_curve = stacked.std(axis=0)

            plt.plot(rounds, mean_curve, label=strategy)
        else:
            plt.plot(rounds, accuracies[0], label=strategy)

    plt.title(f'Accuracy per Active Learning Round - {dataset_name}')
    plt.xlabel('Active Learning Round')
    plt.ylabel('Test Accuracy')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plot_accuracy_curves("MNIST", results)
plot_accuracy_curves("CIFAR10", results)
```

```python
def plot_time_vs_performance(dataset_name, results, figsize=(12,6)):

    dataset = results[dataset_name]

    strategies = list(dataset.keys())

    aubc_vals = []
    time_vals = []

    for strat in strategies:
        aubc_vals.append(dataset[strat]['aubc_mean'])
        time_vals.append(dataset[strat]['time_mean'])

    x = np.arange(len(strategies))
    width = 0.35

    fig, ax1 = plt.subplots(figsize=figsize)

    bars1 = ax1.bar(x - width/2,  aubc_vals, width, label='AUBC (Accuracy)', color='tab:blue')
    ax1.set_ylabel('AUBC (Accuracy)', color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    ax1.set_xticks(x)
    ax1.set_xticklabels(strategies, rotation = 45)


    ax2 = ax1.twinx()
    bars2 = ax2.bar(x + width/2, time_vals, width, label='Runtime (Seconds)', color='tab:orange')
    ax2.set_ylabel('Runtime (Seconds)', color='tab:orange')
    ax2.tick_params(axis='y', labelcolor='tab:orange')

    plt.title(f'Time vs Performance  - {dataset_name}')
    fig.tight_layout()
    plt.show()




```

```python
plot_time_vs_performance("MNIST", results)
plot_time_vs_performance("CIFAR10", results)
```

```python
def plot_final_accuracies(results, dataset_names):

    strategies = list(results[dataset_names[0]].keys())
    num_strategies = len(strategies)
    num_datasets = len(dataset_names)

    x = np.arange(num_strategies)
    width = 0.35

    fig, ax = plt.subplots(figsize=(12,6))

    for i, dataset in enumerate(dataset_names):
        final_accuracies = []

        for strategy in strategies:
            final_accuracies.append(results[dataset][strategy]['accuracies'][0][-1])
        
    
        offset = (i - (num_datasets - 1)/2) * width
        ax.bar(x + offset, final_accuracies, width, label=dataset)

    
    ax.set_xticks(x)
    ax.set_xticklabels(strategies, rotation = 45)
    ax.set_ylabel('Final Accuracy')
    ax.set_title('Final Accuracy per Strategy Across Datasets')
    ax.legend()
    
    plt.tight_layout()
    plt.show()
```

```python
dataset_names = list(results.keys())
plot_final_accuracies(results, dataset_names)
```
